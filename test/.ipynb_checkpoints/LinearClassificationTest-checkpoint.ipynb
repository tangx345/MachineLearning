```python
import sys, os
sys.path.append('F:\\python\\ml\\MachineLearning\\src')
import LinearData 
import LinearClassification
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 18})
```

```python
filename1 = 'LinearData_2D_Gaussian_1'
generator = LinearData.Generator2D_Gaussian([[10, 10], [15, 12], [12, 16]], [[[5, 0], [0, 5]], [[5, 3], [3, 5]], [[5, -3], [-3, 5]]], 100, filename1)
if not os.path.exists('../data/'+filename1+'.csv'):
    generator.generate()
```

```python
data = pd.read_csv('../data/'+filename1+'.csv')
x_cols = data.columns[:-1]
X = data[x_cols].values
Y = data['Y'].values
if len(X.shape) == 1:
    N = X.shape[0]
    X = X.reshape((N, 1))
```

# raw data view

```python
colors = ['r', 'g', 'b']
m = len(data['Y'].unique())

plt.figure()
for i in range(m):
    cur_X = data[data['Y'] == i][x_cols]
    plt.plot(cur_X['x0'], cur_X['x1'], colors[i] + 'o')
```

# LDA

```python
classifier1 = LinearClassification.LDA(data)
y_pred1 = classifier1.predict(X)
df1 = data.copy()
df1['Y'] = y_pred1
```

```python
plt.figure(figsize = (20, 8))
ax1 = plt.subplot(1,2,1)
for i in range(m):
    cur_X = data[data['Y'] == i][x_cols]
    ax1.plot(cur_X['x0'], cur_X['x1'], colors[i] + 'o')
ax2 = plt.subplot(1,2,2)
for i in range(m):
    cur_X = df1[df1['Y'] == i][x_cols]
    ax2.plot(cur_X['x0'], cur_X['x1'], colors[i] + 'o')
```

# QDA

```python
classifier2 = LinearClassification.QDA(data)
y_pred2 = classifier2.predict(X)
df2 = data.copy()
df2['Y'] = y_pred2
```

```python
plt.figure(figsize = (20, 8))
ax1 = plt.subplot(1,2,1)
for i in range(m):
    cur_X = data[data['Y'] == i][x_cols]
    ax1.plot(cur_X['x0'], cur_X['x1'], colors[i] + 'o')
ax2 = plt.subplot(1,2,2)
for i in range(m):
    cur_X = df2[df2['Y'] == i][x_cols]
    ax2.plot(cur_X['x0'], cur_X['x1'], colors[i] + 'o')
```
